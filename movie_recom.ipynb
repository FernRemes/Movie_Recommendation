```python
import numpy as np
import pandas as pd
import difflib

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-Processing

```python
# loading the data from the csv file 
movie_data = pd.read_csv('movies.csv')

# printing the first 5 rows of the dataset
movie_data.head()
```

## Get the Number of rows and columns in the dataset

```python
movie_data.shape
```

```python
# Selecting the relevant features for recommendation

selected_features = ['genres', 'keywords', 'original_title', 'overview', 'tagline', 'title', 'cast', 'director']

print(selected_features)
```

```python
# replacing the null values with a null string

for feature in selected_features:
    movie_data[feature] = movie_data[feature].fillna('')
```

```python
# combining the selected features 

combined_features = movie_data['genres'] + ' ' + movie_data['keywords'] + ' ' + movie_data['original_title'] + ' ' + movie_data['overview'] + ' ' + movie_data['tagline'] + ' ' + movie_data['title'] + ' ' + movie_data['cast'] + ' ' + movie_data['director']

print(combined_features)
```

Converting the text data to feature vectors


```python
vectorizer = TfidfVectorizer()

features_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(features_vectors)
```

Cosine Similarity Algorithm

```python
# Getting the similarity scores between the features using cosine similarity

similarity = cosine_similarity(features_vectors)

print(similarity)

print("rows and column: ", similarity.shape)
```

```python
# Getting the movie name from the user

movie_name = input('Enter Your Favorite Movie Name: ')

print(movie_name)
```

```python
# Creating a list with all the movie names given in the input

list_of_all_titles = movie_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

print(find_close_match)
```

```python
close_match = find_close_match[0]

print(close_match)
```

```python
# Finding the index with title

index_of_movie = movie_data[movie_data.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_movie]))

print(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key=lambda x: x[1], reverse=True)

print(sorted_similar_movies)
```
